```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data Preparation
items_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/Items.csv")

customers_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/customers.csv")

order_item_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/order_item.csv")

orders_csv = pd.read_csv("https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/orders.csv")

#display(customers_csv)
```

```python
# Q2
# Data Analysis
# 1
print("Average price of an item:", items_csv["item_price"].mean())
```

```python
# 2
top_id = order_item_csv.merge(orders_csv, left_on="order_id", right_on="order_id").groupby("customer_id")["quantity"].sum().idxmax()
print(customers_csv[customers_csv["id"] == top_id]["first_name"].values[0])
```
